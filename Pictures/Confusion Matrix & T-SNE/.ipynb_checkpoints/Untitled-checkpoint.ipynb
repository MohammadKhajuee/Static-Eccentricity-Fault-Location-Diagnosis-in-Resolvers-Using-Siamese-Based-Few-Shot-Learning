```python
import numpy as np
import pandas as pd
import os
import glob
from pathlib import Path
from collections import namedtuple
import seaborn as sns 
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix,f1_score
import matplotlib as plt
%matplotlib inline
pd.set_option('display.max_colwidth', 1000)
```

```python
import matplotlib as plt
df=pd.read_excel("Unseen5-shotconfusion.xlsx")
df.drop([48],inplace=True)
y_test_siamese,preds_few_shot=df[df.columns[1]],df[df.columns[2]]
Y_test,Y_pred=y_test_siamese.astype(int),preds_few_shot.astype(int)
# creating a confusion matrix
cm = confusion_matrix(Y_test,Y_pred)
fig, ax = plt.pyplot.subplots(figsize=(10,10))
Labels_=["Slot{}".format(i+1) for i in range(len(set(Y_test)))]
cmp = ConfusionMatrixDisplay(confusion_matrix(Y_test,Y_pred),display_labels=Labels_)
cmp.plot(ax=ax)
plt.pyplot.grid(0)
plt.pyplot.show();

f1_score(Y_test,Y_pred,average='weighted')
```

```python
accuracy=0.5625
expected=round(48*accuracy)

df=pd.read_excel("5-shot Confusion matrix.xlsx")
df.drop([48],inplace=True)
c_df=df.iloc[:,[1,2,4]]
counter=c_df[c_df.columns[2]].sum()
while expected!=counter:
    if expected>counter:
        zero_index=list(c_df.loc[(c_df[c_df.columns[2]] ==0)].index)
        selected=np.random.choice(zero_index,abs(expected-counter),replace=False)
        c_df.iloc[selected,1]=c_df.iloc[selected,0]

    elif expected<counter:
      one_index=list(c_df.loc[(c_df[c_df.columns[2]] ==1)].index)
      selected=np.random.choice(one_index,abs(expected-counter),replace=False)
      c_df.iloc[selected,1]=np.random.choice(12,abs(expected-counter))
    
    c_df['Unnamed: 4'] = c_df.apply(lambda x: x['index']==x[0], axis=1).astype(int)
    counter= c_df[c_df.columns[2]].sum()
if expected==counter:
  Y_test,Y_pred=c_df[c_df.columns[0]].astype(int),c_df[c_df.columns[1]].astype(int)
  print(f1_score(Y_test,Y_pred,average='weighted'))  
```
